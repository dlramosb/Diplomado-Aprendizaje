##### Descripción del Dataset

Información sacada de [UCI Machine Learning Repository]("https://archive.ics.uci.edu/ml/datasets/Beijing+PM2.5+Data#"). El dataset contiene 43824 registros, entre las fechas 1 de Enero de 2010 al 31 de Diciembre de 2014.  Este conjunto de datos (con frecuencia horaria) contiene los datos de PM2.5 de la Embajada de EE. UU 
en Beijing. PM2.5 son los <i style="color:#970D3D">materiales particulados contaminantes presentes en la atmósfera</i> de tamaño igual o menor de 2,5 $\mu$. A continuación se presenta la descripción de las variables contenidas en el conjunto de datos.

- <i style="color:#970D3D">No</i>: Número de Fila.
- <i style="color:#970D3D">year</i>: Año del dato en esta fila.
- <i style="color:#970D3D">month</i>: Mes del dato en esta fila.
- <i style="color:#970D3D">day</i>: Dia del dato en esta fila.
- <i style="color:#970D3D">hour</i>: Hora del dato en esta fila.
- <i style="color:#970D3D">pm2.5</i>: PM2.5 concentracion. (ug/m^3)
- <i style="color:#970D3D">DEWP</i>: Punto de Rocio. 
- <i style="color:#970D3D">TEMP</i>: Temperature.
- <i style="color:#970D3D">PRES</i>: Presión. (hPa)
- <i style="color:#970D3D">cbwd</i>: Dirección combinada del viento.
- <i style="color:#970D3D">Iws</i>: Velocidad del viento acumulada. (m/s)
- <i style="color:#970D3D">Is</i>: Horas acumuladas de nieve.
- <i style="color:#970D3D">Ir</i>: Horas acumuladas de LLuvia.

> 4.Realice un gráfico de dispersión de las variables "pm2.5" y "TEMP" con título, labels apropiados y color rojo para los datos. Describa: ¿Qué se puede inferir de dicho gráfico?

```python
fig=plt.figure(figsize=(12,7))
fig=sns.scatterplot(x=data_beijing["TEMP"],y=data_beijing["pm2.5"],color="red",legend="auto")
fig.set_title("Gráfico de dispersión de variables Temperatura vs Material particulado.")
fig.set_xlabel("Temperatura en [°C]")
fig.set_ylabel("Material particulado pm2.5")
```

Se puede inferir que en temperaturas extremas en beijing la cantidad de material particulado <i>desciende.</i> Ademas que la cantidad de material particulado parece seguir una distribución normal de forma agregada a través de los años.

> 6. Para finalizar, agrupe por año el valor de la variable "pm2.5", calcule el promedio y ordene la base de datos de manera ascendente por el promedio.

```python
agrupados_media=data_beijing[["year","pm2.5"]].groupby(["year"]).mean()
```

```python
agrupados_media
```

El año con menor contaminación fue 2012.

```python
data_beijing[["year","pm2.5"]].groupby(["year"]).describe()
```

> <p style="color:#970D3D">Lo que podemos apreciar es que la distribución para la cantidad de material particulado por año, como lo mostraron los boxplot arriba,  es fuertemente asimetrica hacia la izquierda, y con una cantidad de datos atipicos fuertes, ya que la mediana y la media son diferentes y la media esta a la derecha de la mediana. Esto puede ser influencia principalmente de la velocidad del viento, ya que parece este el factor que influye en el nivel de material particulado presente en la atmósfera.</p>

##### Ordenar la base de datos por el promedio ascendente de pm2.5

```python
data_con_promedio=pd.merge(data_beijing,agrupados_media,how="left",left_on="year",right_on="year")
```

```python
data_con_promedio.rename(columns={"pm2.5_y":"pm2.5_promedio"}).set_index(["year","pm2.5_promedio"]).sort_values(ascending=True,by=["pm2.5_promedio","Fecha"])
```
